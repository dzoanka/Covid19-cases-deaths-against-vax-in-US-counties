### Population density

```python
df_pop = pd.read_csv('input_data/covid_county_population_usafacts.csv')

pop_ski = list()
for x in fips_ski:
    pop_ski.append(df_pop.loc[df_pop['countyFIPS'] == x]['population'].item())

pop_uni = list()
for x in fips_uni:
    pop_uni.append(df_pop.loc[df_pop['countyFIPS'] == x]['population'].item())
```

```python
df_area = pd.read_csv('input_data/county_complete.csv')

area_ski = list()
for fips in fips_ski:
    area_ski.append(df_area.loc[df_area['fips'] == fips]['area_2010'].item())
    
area_uni = list()
for fips in fips_uni:
    area_uni.append(df_area.loc[df_area['fips'] == fips]['area_2010'].item())
```

```python
dens_ski = [round(pop / area_ski[idx], 2) for idx, pop in enumerate(pop_ski)]
dens_uni = [round(pop / area_uni[idx], 2) for idx, pop in enumerate(pop_uni)]
```

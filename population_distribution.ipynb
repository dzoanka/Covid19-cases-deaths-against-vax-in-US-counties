## This notebook presents the `distribution of population` and `distribution of population density` across US counties.

```python
# Filename where it is read from: must be used the file from readable_data_table.ipynp processing
filename = "counties_data_jan22.csv"
```

```python
#Imports
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(filename)
```

## Population distribution

```python
print('Minimum population: ', df['Population'].min())
print('Maximum population: ', df['Population'].max())
```

```python
sns.displot(df, x="Population")
```

```python
df_smaller_2mln =  df[ df['Population'] < 2000000]
sns.histplot(df_smaller_2mln, x="Population")
```

```python
df_smaller_1mln =  df[ df['Population'] < 1000000] 
sns.histplot(df_smaller_1mln, x="Population")
```

```python
df_smaller_500k =  df[ df['Population'] < 500000] 
sns.histplot(df_smaller_500k, x="Population")
```

```python
df_smaller_200k =  df[ df['Population'] < 200000] 
g = sns.histplot(df_smaller_200k, x="Population")
xlabels = ['{:,.0f}'.format(x) + 'k' for x in g.get_xticks()/1000]
g.set_xticklabels(xlabels)
```

```python
print("Counties above 200k inhabitants: ", len(df[ df['Population'] > 200000] ))
print("Counties below 200k inhabitants: ", len(df[ df['Population'] < 200000] ))
```

```python
print("Counties above 25k inhabitants: ", len(df[ df['Population'] > 25000] ))
print("Counties below 25k inhabitants: ", len(df[ df['Population'] < 25000] ))
```

## Population density distribution

```python
print('Minimum population density: ', df['Pop. Density'].min())
print('Maximum population density: ', df['Pop. Density'].max())
```

```python
sns.displot(df, x="Pop. Density")
```

```python
df_smaller_10k =  df[ df['Pop. Density'] < 10000] 
sns.histplot(df_smaller_10k, x="Pop. Density")
```

```python
df_smaller_2k =  df[ df['Pop. Density'] < 2000] 
sns.histplot(df_smaller_2k, x="Pop. Density")
```

```python
df_smaller_500 =  df[ df['Pop. Density'] < 500] 
sns.histplot(df_smaller_500, x="Pop. Density")
```

```python
print("Counties with density above 500 inhabitants per square mile: ", len(df[ df['Pop. Density'] > 500] ))
print("Counties with density above 500 inhabitants per square mile: ", len(df[ df['Pop. Density'] < 500] ))
```

```python
df_smaller_200 =  df[ df['Pop. Density'] < 200] 
sns.histplot(df_smaller_200, x="Pop. Density")
```
